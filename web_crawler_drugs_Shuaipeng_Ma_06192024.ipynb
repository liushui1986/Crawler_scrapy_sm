## 1. Get webpage using *requests*

```python
import requests

req = requests.get('https://en.wikipedia.org/wiki/Drug')
```

```python
webpage = req.text
```

```python
print(webpage)
```

## 2. Get specific contents using BeatifulSoup

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(webpage, 'html.parser')
```

```python
soup
```

### 2.1 Prettify the webpage

```python
print(soup.prettify())
```

### 2.2 Get all paragraphs

```python
paragraphs = soup.find_all('p', attrs={"class":False})
```

```python
paragraphs
```

### 2.3 Get all the links in these paragraphs which point to other webpages

```python
# if 'paragraph' is a single element, we can use 
# paragraph.find_all('a')
# find_all will find all
# find will only find the first one
# if 'paragraphs' is a 'ResultSet' (a list of elements), we can interate over each paragraph to find the first link in each
for para in paragraphs:
    links = para.find_all('a', attrs={"title":True})
    for link in links:
        print(link)
```

```python
len(links)
```

```python
# Data dictionary to store titles and hrefs
data = {"title":[], "link":[]}

head = "https://en.wikipedia.org"
# Iterate over each paragraph
for para in paragraphs:
    # Find all 'a' tags with a title attribute within the current paragraph
    for link in para.find_all('a', attrs={"title":True}):
        data["title"].append(link["title"])
        data["link"].append(head + link["href"])
```

```python
import pandas as pd
df = pd.DataFrame(data)
```
