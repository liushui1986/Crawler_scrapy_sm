```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import csv
import time
from selenium.webdriver.chrome.service import Service

# Set up the Chrome WebDriver
s = Service('/Users/mashu/Downloads/chromedriver-win64/chromedriver.exe')

driver = webdriver.Chrome(service = s)

# URL of the LinkedIn job listings page and Open the URL
driver.get("https://www.linkedin.com/jobs/search?keywords=Data%20Scientist&location=United%20States&geoId=103644278&f_TPR=&f_WT=2&position=1&pageNum=0")

# Allow the page to load
time.sleep(5)  # Adjust sleep time as necessary

scroll_pause_time = 10  # Wait time between scrolls
scroll_height = 0

while True:
    # Get current scroll height
    prev_scroll_height = scroll_height
    scroll_height = driver.execute_script("return document.body.scrollHeight")
    
    # Scroll down to the bottom
    driver.execute_script(f"window.scrollTo(0, {scroll_height});")

    # Wait to load the page
    time.sleep(scroll_pause_time)  # Adjust sleep time as necessary

    # Calculate new scroll height and compare with last scroll height
    if scroll_height == prev_scroll_height:
        break

# Locate the job listings container and find all job listings
job_listings = driver.find_elements(By.XPATH, '//*[@id="main-content"]/section[2]/ul/li')

# Initialize lists to store data
job_titles = []
company_names = []
addresses = []
job_links = []

# Extract and print the job titles
for listing in job_listings:
    job_title_element = listing.find_element(By.CLASS_NAME, 'sr-only')
    job_title = job_title_element.text
    job_titles.append(job_title)

    company_name_element = listing.find_element(By.CLASS_NAME, 'base-search-card__subtitle')
    company_name = company_name_element.text
    company_names.append(company_name)

    address_element = listing.find_element(By.CLASS_NAME, 'base-search-card__metadata')
    address = address_element.text
    addresses.append(address)

    job_link_element = listing.find_element(By.CLASS_NAME, 'base-card__full-link')
    job_link = job_link_element.get_attribute('href')
    job_links.append(job_link)

    print('Job Title: ', job_title)
    print('Company Name: ', company_name)
    print('Address: ', address)
    print('Job Link: ', job_link)
    print('-' * 50)

# Close the WebDriver
driver.quit()
```

```python
# Create a DataFrame with the scraped data
df = pd.DataFrame({
    'Job Title': job_titles,
    'Company Name': company_names,
    'Address': addresses,
    'Job Link': job_links
})

# Save the DataFrame to a CSV file
df.to_csv('linkedin_job_listings_new.csv', index=False)
```
